```python
import string
import re
from numpy import array, argmax, random, take
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense, LSTM, Embedding, RepeatVector
from keras.preprocessing.text import Tokenizer
from keras.callbacks import ModelCheckpoint
from keras.preprocessing.sequence import pad_sequences
from keras.models import load_model
from keras import optimizers
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def read_file(filename) -> string:
    file = open(filename, 'r', encoding='utf-8')
    text = file.read()
    file.close()
    return text
```

```python
def split_to_sentence(text):
    sents = text.strip().split('\n')
    sents = [i.split('\t') for i in sents]
    return sents
```

```python
data = read_file("C:/deu.txt")
deu_eng = split_to_sentence(data)
deu_eng = array(deu_eng)
```

```python
deu_eng = deu_eng[:50000,:]
```

```python
def remove_punctuation(text):
    translator = str.maketrans('','', string.punctuation)
    return text.translate(translator)
```

```python
print(remove_punctuation(deu_eng[0,1]))
```

```python
for sentence in deu_eng:
    for word in [0,1]:
        sentence[word] = remove_punctuation(sentence[word])
```

```python
for sentence in deu_eng:
    for word in [0,1]:
        sentence[word] = sentence[word].lower()
```

```python
eng_l = []
deu_l = []
```

```python
for i in deu_eng[:,0]:
      eng_l.append(len(i.split()))

for i in deu_eng[:,1]:
      deu_l.append(len(i.split()))

length_df = pd.DataFrame({'eng':eng_l, 'deu':deu_l})
```

```python
def tokenization(text):
    tokenizer = Tokenizer()
    tokenizer.fit_on_texts(text)
    return tokenizer
```

```python
eng_t = tokenization(deu_eng[:, 0])
eng_vocab_len = len(eng_t.word_index) + 1
print(f"English vocabulary size is {eng_vocab_len}")
```

```python
deu_tokenizer = tokenization(deu_eng[:, 1])
deu_vocab_size = len(deu_tokenizer.word_index) + 1


eng_tokenizer = tokenization(deu_eng[:, 0])
eng_vocab_size = len(eng_tokenizer.word_index) + 1

eng_length = 8
deu_length = 8
print('Deutch vocabulary Size: %d' % deu_vocab_size)
```

```python
print(eng_tokenizer.word_index)
```

```python
def encode_sentence(tokenizer, size, text):
    seq = tokenizer.texts_to_sequences(text)
    seq = pad_sequences(seq, maxlen=size, padding = 'post')
    return seq
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(deu_eng, test_size = 0.2, random_state = 12)
```

```python
print(train)
```

```python
x_train = encode_sentence(deu_tokenizer, deu_length, train[:, 1])
y_train = encode_sentence(eng_tokenizer, eng_length, train[:, 0])
```

```python
x_test = encode_sentence(deu_tokenizer, deu_length, test[:, 1])
y_test = encode_sentence(eng_tokenizer, eng_length, test[:, 0])
```

```python
print(x_test, x_test.shape)
```

```python
model = Sequential()
model.add(Embedding(deu_vocab_size, 256, input_length=deu_length))
model.add(LSTM(512))
model.add(RepeatVector(eng_length))
model.add(LSTM(512, return_sequences=True))
model.add(Dense(eng_vocab_size, activation='softmax'))
print(model.summary())
```

```python
rms = optimizers.RMSprop(learning_rate=0.001)
model.compile(optimizer=rms, loss='sparse_categorical_crossentropy')
```

```python
history = model.fit(x_train,  y_train, epochs=30, batch_size=512, validation_split = 0.2)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.legend(['train','validation'])
plt.show()
```

```python
preds = model.predict(x_test.reshape((x_test.shape[0],x_test.shape[1])))
```

```python
def get_word(n, tokenizer):
    for word, index in tokenizer.word_index.items():
        if np.all(index == n):
            return word
    return None
```

```python
preds_text = []
for i in preds:
    temp = []
    for j in range(len(i)):
        t = get_word(i[j], eng_tokenizer)
        if j > 0:
            if (t == get_word(i[j-1], eng_tokenizer)) or (t == None):
                temp.append('')
            else:
                temp.append(t)
        else:
            if(t == None):
                temp.append('')
            else:
                temp.append(t) 

    preds_text.append(' '.join(temp))
```

```python
pred_df = pd.DataFrame({'actual' : test[:,0], 'predicted' : preds_text})
```
